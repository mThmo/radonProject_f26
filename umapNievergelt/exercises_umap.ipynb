### 1.)
A yet unknown point $(w_1, w_2)$ has orthogonal projections $2.284$ and $1.812$ on the lines supporting the vectors $(0.936, 0.352)$ and $(-0.352, 0.936)$. Determine $w_1$ and $w_2$.

Setting up a matrix to represent the system of equations, 

\begin{align*}
0.936 w_1 + 0.352 w_2 &= 2.284 \\
-0.352 w_1 + 0.936 w_2 &= 1.812 \\
\end{align*}

\begin{align*}
\begin{pmatrix} 0.936 & 0.352 \\ -0.352 & 0.936 \end{pmatrix}
\begin{pmatrix} w_1 \\ w_2 \end{pmatrix} =
\begin{pmatrix} 2.284 \\ 1.812 \end{pmatrix}
\end{align*}

\begin{align*} 
((0.936 / 0.352)0.936 + 0.352) w_2 &= ((0.936 / 0.352)1.812 + 2.284) \\
w_2 &= ((0.936 / 0.352)1.812 + 2.284) / ((0.936 / 0.352)0.936 + 0.352) \\
w_2 &= 5/2 \\
w_1 &= (2.284 - (0.352)(5/2))/0.936 \\
w_1 &= 3/2 \\
\implies \begin{pmatrix} w_1 \\ w_2 \end{pmatrix} &= \begin{pmatrix} 3/2 \\ 5/2 \end{pmatrix}
\end{align*}
